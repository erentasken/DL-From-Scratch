```python
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.stats import norm

# Parameters for the normal distribution
mean = 0      # mean (mu)
std_dev = math.sqrt( 1 / 135)   # standard deviation (sigma)

# Generate values
x = np.linspace(-5, 5, 1000)  # 1000 points from -5 to 5
y = norm.pdf(x, mean, std_dev)  # probability density function
# Count values

# Plot
plt.plot(x, y, label=f'Normal Distribution\nmean={mean}, std={std_dev}')
plt.fill_between(x, y, alpha=0.2)
plt.title('Normal Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```
